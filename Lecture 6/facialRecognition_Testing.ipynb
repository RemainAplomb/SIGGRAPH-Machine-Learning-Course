```
import mtcnn
mtcnn_model = mtcnn.MTCNN()
```

```
# download an image
!curl -o issth.jpeg https://i.gr-assets.com/images/S/compressed.photo.goodreads.com/books/1498315806l/35507089._SX318_.jpg
art_faces = cv2.imread('issth.jpeg')
cv2_imshow(art_faces)
```

```
def recognize_faces(new_img):
  faces = mtcnn_model.detect_faces(new_img)
  for i in range(0, len(faces)):
      x, y, w, h = faces[i]['box']
      # draw rectangle
      cv2.rectangle(new_img, (x, y), (x+w, y+h), (255, 0, 0), 1)
  return new_img
```

```
img = art_faces
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
final_img = recognize_faces(img)
cv2_imshow(final_img[:,:,::-1])
```
